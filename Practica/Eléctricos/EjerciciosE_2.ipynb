## Ejercicio 2

La Multinacional Vishay desea implementar un algoritmo que permita determinar el valor de una resistencia eléctrica basados en el código de colores (ver figura 1). El algoritmo debe solicitar a usuario el color en MAYUSCULA de cada una de las tres bandas (en caso de un color incorrecto mostrar el siguiente mensaje al usuario “introduzca un color valido”) y mostrar el valor de la resistencia en Ohms con sus respectivos colores en cada banda.
